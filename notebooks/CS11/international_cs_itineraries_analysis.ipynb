Some itineraries have no total emission but they should have. I will check if these are only international flights

```python
def contains_numbers(str):
    for char in str:
        if char.isdigit():
            return True
    return False
```

```python
for idx,row in possible_itineraries_clustered_pareto[(possible_itineraries_clustered_pareto["total_emissions"].isna())&(~possible_itineraries_clustered_pareto["journey_type"].str.contains("none"))].iterrows():
    if contains_numbers(row["origin"]):
        if contains_numbers(row["destination"]):
            print(f"there is a national trip without CO2 at index {idx}")


print("done!")
```
